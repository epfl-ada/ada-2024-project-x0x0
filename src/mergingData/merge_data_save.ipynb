```python
def get_var_name(var):
    for name, value in globals().items():
        if value is var:
            return name
```

```python
def save_csv_with_directory_check(df, file_path):
    directory = os.path.dirname(file_path)
    
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    df.to_csv(file_path, index=False)
    print(f"CSV saved to {file_path}")
```

```python
save_csv_with_directory_check(BA_beer_loc, 'data/baseData/Modified/' + get_var_name(BA_beer_loc) + '.csv')
save_csv_with_directory_check(BA_rating_loc, 'data/baseData/Modified/' + get_var_name(BA_rating_loc) + '.csv')
save_csv_with_directory_check(RB_beer_loc, 'data/baseData/Modified/' + get_var_name(RB_beer_loc) + '.csv')
save_csv_with_directory_check(RB_rating_loc, 'data/baseData/Modified/' + get_var_name(RB_rating_loc) + '.csv')
save_csv_with_directory_check(BA_ratbeer, 'data/baseData/Modified/' + get_var_name(BA_ratbeer) + '.csv')
save_csv_with_directory_check(RB_ratbeer, 'data/baseData/Modified/' + get_var_name(RB_ratbeer) + '.csv')
```
