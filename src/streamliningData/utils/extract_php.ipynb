```python

import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url = "https://state.1keydata.com/bordering-states-list.php"
```

```python
r = requests.get(url)

print('Response status code: {0}\n'.format(r.status_code))
print('Response headers: {0}\n'.format(r.headers))
print('Response body: {0}'.format(r.text))
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
with open("bordering_states.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["State", "Bordering States"])  # CSV headers

    # Loop through each list item containing state and borders
    for state_section in soup.find_all("li"):
        # Split state from its bordering states
        if ":" in state_section.text:
            state, borders = state_section.text.split(":", 1)
            writer.writerow([state.strip(), borders.strip()])

print("CSV file 'bordering_states.csv' created successfully.")
```
